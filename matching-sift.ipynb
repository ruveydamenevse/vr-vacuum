```python
 # Brute-Force Matching with SIFT Descriptors
```

```python
img1 = cv2.imread('image0.jpeg',cv2.IMREAD_GRAYSCALE) # query
img2 = cv2.imread('image1.jpeg',cv2.IMREAD_GRAYSCALE) # train
```

```python
sift = cv2.SIFT_create() #open a SIFT
```

```python
#keypoints and descriptors with SIFT
```

```python
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)
```

```python
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)
```

```python
# ratio test by D.Lowe
```

```python
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])
```

```python
# list of lists as matches
```

```python
img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
plt.imshow(img3)
plt.show()
```
